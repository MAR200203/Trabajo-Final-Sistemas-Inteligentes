```python
import pandas as pd

# cargar los archivos csv con las caracteristicas de hogares y viviendas de los años 2021, 2022 y 2023

df_caracteristicas_hogar_2021 = pd.read_csv('C:/Users/ampar/OneDrive/Desktop/Trabajo-Final-Sistemas-Inteligentes/MPI_2021_2023/Dataset_MPI_2021_2023/caracteristicas_hogar_2021_759-Modulo01.csv', sep=',', encoding='latin1')
df_caracteristicas_hogar_2022 = pd.read_csv('C:/Users/ampar/OneDrive/Desktop/Trabajo-Final-Sistemas-Inteligentes/MPI_2021_2023/Dataset_MPI_2021_2023/caracteristicas_hogar_2022_784-Modulo01.csv', sep=',', encoding='latin1')
df_caracteristicas_hogar_2023 = pd.read_csv('C:/Users/ampar/OneDrive/Desktop/Trabajo-Final-Sistemas-Inteligentes/MPI_2021_2023/Dataset_MPI_2021_2023/caracteristicas_hogar_2023_906-Modulo01.csv', sep=',', encoding='latin1')

# caracteristicas_hogar_2021

print("-"*50)
print("caracteristicas de hogares y viviendas 2021")
df_caracteristicas_hogar_2021.info()

# caracteristicas_hogar_2022

print("-"*50)
print("caracteristicas de hogares y viviendas 2022")
df_caracteristicas_hogar_2022.info()

# caracteristicas_hogar_2023

print("-"*50)
print("caracteristicas de hogares y viviendas 2023")
df_caracteristicas_hogar_2023.info()

# obtener los nombres de las columnas

columnas_caracteristicas_hogar_2021 = df_caracteristicas_hogar_2021.columns
columnas_caracteristicas_hogar_2022 = df_caracteristicas_hogar_2022.columns
columnas_caracteristicas_hogar_2023 = df_caracteristicas_hogar_2023.columns


print("-"*50)

print("columas con coincidencias en los tres años")

# obtener las columnas que coinciden en los tres años

columnas_coincidentes = list(set(columnas_caracteristicas_hogar_2021) & set(columnas_caracteristicas_hogar_2022) & set(columnas_caracteristicas_hogar_2023) )
print(columnas_coincidentes)
print(len(columnas_coincidentes))

# obtener las columnas que no coinciden en los tres años

print("-"*50)
columnas_no_coincidentes_2021 = list(set(columnas_caracteristicas_hogar_2021) - set(columnas_caracteristicas_hogar_2022) - set(columnas_caracteristicas_hogar_2023) )
print(columnas_no_coincidentes_2021)
print(len(columnas_no_coincidentes_2021))

print("-"*50)
columnas_no_coincidentes_2022 = list(set(columnas_caracteristicas_hogar_2022) - set(columnas_caracteristicas_hogar_2021) - set(columnas_caracteristicas_hogar_2023) )
print(columnas_no_coincidentes_2022)
print(len(columnas_no_coincidentes_2022))

print("-"*50)
columnas_no_coincidentes_2023 = list(set(columnas_caracteristicas_hogar_2023) - set(columnas_caracteristicas_hogar_2021) - set(columnas_caracteristicas_hogar_2022) )
print(columnas_no_coincidentes_2023)
print(len(columnas_no_coincidentes_2023))

# unir los dataframes y generar un archivo csv con los datos consolidados de los tres años con las columnas coincidentes

df_caracteristicas_hogar_2021 = df_caracteristicas_hogar_2021[columnas_coincidentes]
df_caracteristicas_hogar_2022 = df_caracteristicas_hogar_2022[columnas_coincidentes]
df_caracteristicas_hogar_2023 = df_caracteristicas_hogar_2023[columnas_coincidentes]

df_caracteristicas_hogar = pd.concat([df_caracteristicas_hogar_2021, df_caracteristicas_hogar_2022, df_caracteristicas_hogar_2023])

df_caracteristicas_hogar.to_csv('C:/Users/ampar/OneDrive/Desktop/Trabajo-Final-Sistemas-Inteligentes/MPI_2021_2023/Dataset_MPI_2021_2023/caracteristicas_hogar_2021_2023.csv', index=False, encoding='latin1')

```
